```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
plt.style.use('ggplot')
```

```python
dismissal_types = ['c', 'b', 'lbw', 'st', 'run out']
pattern = '|'.join(dismissal_types)  # Creates a pattern like 'caught|bowled|lbw|stumped|run out'

df['dismissal_type'] = df['dismissal'].str.extract('('+ pattern + ')', expand=False)
df['dismissal_type']
```

```python
df['dismissal_type'].isna().sum()
```

# üèè Exploratory Data Analysis and Visualizations

1. Who are the top 10 players with the highest runs?

```python
# extract the top players
top_players = df.nlargest(10, 'runs')

# create the bar plot
plt.figure(figsize=(10,6))
sns.barplot(data=top_players, x='runs', y='player', palette='viridis_r')
plt.title('TOP 10 PLAYERS WITH HIGHEST RUNS', size='15', fontweight='bold')
plt.xlabel('Runs')
plt.ylabel('Player Name')
plt.show()
```

2. What is the distribution of runs across top players?

```python
top_table = top_players[['player','runs']].reset_index(drop=True)
top_table
```

```python
# plot the histogram
plt.figure(figsize=(8,6))
sns.histplot(data=top_players, x='runs', y='player', color='blue')
plt.title('DISTRIBUTION OF RUNS ACROSS TOP PLAYERS', size='15', fontweight='bold')
plt.xlabel('Runs')
plt.ylabel('Player Name')
plt.show()
```

3. Who are the players with the most 4s and 6s?

```python
# Create a new column of 4s and 6s
df['fours_sixes'] = df['4s'] + df['6s']

# Create barplot
plt.figure(figsize=(8,6))
sns.barplot(y='player', x='fours_sixes', data=df.nlargest(10, 'runs'), palette='RdYlGn_r')
plt.title('PLAYERS WITH MOST 4S AND 6S', size='15', fontweight='bold')
plt.xlabel('Runs')
plt.ylabel('Player Name')
plt.show()
```

4. What is the correlation between runs and balls?

```python
# find correlation
correlation = df[['runs','balls']].corr()

# plot heatmap
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('CORRELATION BETWEEN RUNS AND BALLLS')
plt.show()
```

- **correlation**: A correlation of `0.93` between `runs` and `balls` in cricket indicates a strong positive relationship. This means that as the number of balls faced by a player increases, the number of runs scored also tends to increase.

5. What is the average number of runs per dismissal type?

```python
# Create barplot
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='dismissal_type', y='runs', ci=None, palette='viridis')
plt.title('AVERAGE RUNS PER DISMISSAL TYPE', fontweight='bold')
plt.ylabel('Runs')
plt.xlabel('Dismissal Type')
plt.show()
```

- **Distribution**: The graph shows the distribution of average runs across different types of dismissals. The ‚Äò`c`‚Äô dismissal type i.e `Caught Out` has the highest average runs, while the ‚Äò`st`‚Äô dismissal type i.e `Stumped` has the lowest.

6. What is the distribution of minutes played by players?

```python
plt.figure(figsize=(12,6))
sns.histplot(data=df, x='minutes')
plt.title('DISTRIBUTION OF MINUTES PLAYED BY PLAYERS', fontweight='bold')
plt.xlabel('Minutes')
plt.ylabel('Count')
plt.show()
```

- The chart shows the distribution of minutes played by players. The x-axis represents the ‚ÄúMinutes‚Äù ranging from 0 to 200, and the y-axis represents the ‚ÄúCount‚Äù ranging from 0 to 200. The bars decrease in height as the minutes increase, indicating that fewer players play for longer durations. The highest bar is at 0 minutes with a count of around 175, and the lowest bar is at 200 minutes with a count of around 25. This suggests that most players play for a short duration, while a few players play for longer durations.

7. Who are the top 10 players who faced the most balls?

```python
# Extract the players who faces the most balls
top_balls_faced = df.nlargest(10, 'balls')

balls_faced_table = top_balls_faced[['player', 'balls']].reset_index(drop=True)
balls_faced_table
```

```python
# create the bar plot
plt.figure(figsize=(10,6))
sns.barplot(data=top_balls_faced, x='balls', y='player', palette='viridis')
plt.title('TOP 10 PLAYERS WHO FACED THE MOST BALLS', size='15', fontweight='bold')
plt.xlabel('Balls Faced')
plt.ylabel('Player Name')
plt.show()
```

8. What is the correlation between 4s and 6s?

```python
# Find correlation
cor = df[['4s','6s']].corr()

# Plot heatmap
sns.heatmap(cor, annot=True, cmap='coolwarm')
plt.title('CORRELATION BETWEEN 4S AND 6S')
plt.show()
```

- **correlation**: There is a correlation of `0.53`. This suggests that there is a positive correlation of 0.53 between the number of 4s and 6s scored by players. In other words, players who score more 4s are likely to score more 6s, and vice versa.

9. What are the number of 4s and 6s of top players?

```python
# Melt the DataFrame to have '4s' and '6s' in the same column
melted_df = pd.melt(top_players, id_vars='player', value_vars=['4s', '6s'], 
                    var_name='Type', value_name='Count')

# Create the multi-bar chart
plt.figure(figsize=(20,10))
bar_plot = sns.barplot(x='player', y='Count', hue='Type', data=melted_df)
for p in bar_plot.patches:
    bar_plot.annotate(format(p.get_height(), '.1f'), 
                      (p.get_x() + p.get_width() / 2., p.get_height()), 
                      ha = 'center', va = 'center', 
                      xytext = (0, 9), 
                      textcoords = 'offset points')
plt.title('NUMBER OF 4S AND 6S OF TOP 10 PLAYERS', fontweight='bold', size=20)
plt.xlabel('Player Name', size=12)
plt.ylabel('Count', size=12)
plt.show()
```
