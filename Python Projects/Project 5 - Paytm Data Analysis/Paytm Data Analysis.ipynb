# ðŸ’¸ About the Dataset

This dataset contains Paytm Revenue, Users, Transactions listed operational metrics and revenue of Indian Payment App 'Paytm' from June, 2020 to March, 2023.

Note: All money figures are in INR (Indian National Rupee).

1 Cr is 1 Crore in Indian Number System and 1 Cr is equivalent to 10 million. 1 Lakh is equal to 100 thousand.

## **Data Dictionary**:

| Columns | Description |
| --- | --- |
|Gross Merchandise Value (GMV) :| Value of total payments made to merchants through Paytm app. It is in INR Lac Cr (100,000 Crores).|
|Merchant Transactions :| Total number of Transactions made to merchant through Paytm App. It is in millions.|
|Total Transaction :| Total number of Merchant Transactions and Non-merchant transactions (person to person). It is in millions.|
|Average Monthly Transacting Users (MTU) :| Unique merchants that are listed for receiving a transaction through Paytm QR or a Paytm issued device. It is in millions.|
|Registered Merchants (end of period) :| Number of Registered merchants account in millions.|
|Payment Devices (cumulative, end of period):| Total number of Paytm POS devices and Soundbox deployed at our in-store merchants. It is in millions.|
|Volume of Loan Distributed :| Total volume of loans distributed to users. It is in Thousands.
|Postpaid Loans:| Volume of loans distributed to postpaid users   |
|Personal Loans:| Volume of loans distributed to personal users  | 
|Merchant Loans:|Volume of loans distributed to merchants.|
|Value of loans distributed :| Value of loans distributed to users in INR Cr (Crore).| 
|Postpaid Loans.1 :| Value of loans distributed to postpaid users  | 
|Personal Loans.1 :| Value of loans distributed to personal users  | 
|Merchant Loans.1 :|Value of loans distributed to merchants |
|Payments Services to Consumers:| Revenue generated from consumers in INR crores.| 
|Payments Services to Merchants:| Revenue generated from merchants in INR crores.  |
|Other Operating Revenue:| Revenue generated from other sources in INR crores.|
|Total Payments Revenue: | Total Revenue generated |
|Payment processing charges:| Payment processing charges is charge paid by Paytm (in INR crores).| 
|Net Payments Margin:| It is subtracted from total revenue to get Net Payments margin in INR crores.|

# ðŸ’¸ Exploratory Data Analysis and Visualizations

1. What is the distribution of â€˜Quarterly Merchandise Value Transactionsâ€™?

```python
plt.figure(figsize=(20,10))
sns.barplot(data = df, x='date', y='gross_merc_value', palette='RdYlGn_r')
plt.title('DISTRIBUTION OF QUARTERLY MERCHANDISE VALUE TRANSACTIONS', fontweight='bold', size=20)
plt.xlabel('Date', size=18)
plt.ylabel('Gross Merchandise Value(in Lac Cr)', size=18)
plt.xticks(size=13)
plt.yticks(size=13)
plt.show()
```

- There is a gradual increase in the `Gross Merchandise Value` over every quarter.

2. What is the correlation between â€˜Payment Volumeâ€™ and â€˜Merchant Loansâ€™?

```python
correlation = df[['merchant_trans' ,'merchant_loan_volume']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Merchant Transaction Vs Merchant Loan Volume')
plt.show()
```

- A correlation of `0.93` indicates a very strong positive linear relationship between two variables. It suggests that when the transaction volume for a merchant increases, the loan volume for the merchant also tends to increase, and vice versa.

3. Are there any outliers in the â€˜Payment Processing Chargesâ€™?

```python
# Plot the boxplot of the 'Payment processing charges' column
sns.boxplot(y='pay_processing_charge', data=df)
plt.title('Boxplot of Payment Processing Charges')
plt.show()
```

4. What is the trend of â€˜Loan Volume of Merchantsâ€™ over time?

```python
plt.figure(figsize=(16,8))
sns.lineplot(data=df, x='date', y='merchant_loan_volume')
plt.title('VOLUME TREND OF MERCHANT LOANS', size=20, fontweight='bold')
plt.xlabel('Date', size=15)
plt.ylabel('Merchant loan volume', size=15)
plt.show()
```

- The `merchant loan volume` increases from `01-06-2020` to `01-12-2020`, then there is a decline from `01-12-2020` to `01-06-2021`. From `01-06-2021` there is consistent increase of `merchant loan volume` overall.

5. Which quarter has generated the highest â€˜Revenueâ€™?

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='date', y='total_pr', palette='viridis')
plt.title('HIGHEST TOTAL PAYMENTS REVENUE', size=20, fontweight='bold')
plt.xlabel('Quarter', size=15)
plt.ylabel('Total Payments Revenue (in millions)', size=15)
plt.show()
```

- The Quarter `01-03-2023` generated the highest revenue.

6. Is there a significant difference in â€˜Revenue generated from consumersâ€™ over different quarters?

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='date', y='ps_consumers', palette='mako')
plt.title('QUARTERLY REVENUE GENERATED FROM CONSUMERS', size=20, fontweight='bold')
plt.xlabel('Quarter', size=15)
plt.ylabel('Revenue generated from consumers (in crores)', size=15)
plt.show()
```

- The quarter `01-09-2022` generated the highest revenue from comsumers.

7. What is the trend of â€˜Personal Loanâ€™ value over time?

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=df, x='date', y='personal_loan_value', color='green')
plt.title('TREND OF PERSONAL LOAN VALUE', size=20, fontweight='bold')
plt.xlabel('Quarter', size=15)
plt.ylabel('Personal loan value (in crores)', size=15)
plt.show()
```

- Trend: The green line shows a steady increase in personal loan value over time.

8. What is the correlation between â€˜Payment Revenueâ€™ and â€˜Payment Processing Chargesâ€™?

```python
# find correlation between `total payment revenue` and `payment processing charges`
correlation = df[['total_pr', 'pay_processing_charge']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Total Payment Revenue Vs Payment Processing Charges')
plt.show()
```

- A correlation of `-0.77` between `total payment revenue` and `payment processing charges` indicates a strong negative relationship between these two variables. This means that as `total payment revenue` increases, `payment processing charges` tend to decrease, and vice versa.
