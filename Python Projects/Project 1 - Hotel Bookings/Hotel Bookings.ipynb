# üè®About the Dataset

This dataset contains 119390 observations for a City Hotel and a Resort Hotel. Each observation represents a hotel booking between the 1st of July 2015 and 31st of August 2017, including booking that effectively arrived and booking that were canceled.

Contents:

Since this is hotel real data, all data elements pertaining hotel or costumer identification were deleted. Four Columns, `name`, `email`, `phone number` and `credit_card` have been artificially created and added to the dataset.

# üè®Problem Statement

In the years 2015 to 2017, City Hotel and Resort Hotel have seen high cancellation rates. Each hotel is dealing with a number of issues as a result, including fewer revenues and less than ideal hotel room use. Consequently, lowering cancellation rates is both hotels' primary goal in order to increase their efficiency in generating revenue, and for us to offer thorough business advice to address this problem.

The analysis of hotel booking cancellations as well as other factors that have no bearing on their business and yearly revenue generation are the main topics of this report.

# üè®Assumptions
- No unusual occurrences between 2015 and 2017 will have a substantial impact on the data used.
- The information is still current and can be used to analyze a hotel's possible plans inan efficient manner.
- There are no unanticipated negatives to the hotel employing any advised technique.
- The hotels are not currently using any of the suggested solutions.
- The biggest factor affecting the effectiveness of earning income is booking cancellations.
- Cancellations result in vacant rooms for the booked length of time.
- Clients make hotel reservations the same year they make cancellations.

# üè®Research Questions 

1. What are the factors that are effecting the high hotel reservation cancellations?
2. How can we make hotel reservations cancellations better?
3. How will hotels be assisted in making pricing and promotional decisions?

# üè®Hypothesis

1. More cancellations occur when price is higher.
2. When there is a longer waiting list, customers tend to cancel more frequently.
3. The majority of the clients are coming from offline travel agents to make their reservations.

### Visualizing Average daily rate (adr) of City and Resort hotels.
The Average Daily Rate (ADR) is a key performance indicator that represents the average room rate that a guest pays per room per day. It is calculated by dividing the total room revenue by the total number of rooms sold over a specific period.
![img](https://duve.com/wp-content/uploads/2022/10/adr-formula.jpg)

```python
#Filtereing the data for resort and city hotels separately
resort = resort.groupby('reservation_status_date')[['adr']].mean()
city = city.groupby('reservation_status_date')[['adr']].mean()
```

## Creating an Average Daily Rate Visualization of both City and Resort Hotels

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotels', fontsize = 30)
plt.plot(resort.index, resort['adr'], label = 'Resort Hotel')
plt.plot(city.index, city['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

- The City hotels price seems to fluctuate less in comparison to Resort hotels.
- For the most part City hotel prices are lower than Resort hotels. But some days it is vice-versa.
- Both City and Resort hotel seem to have spikes which could be the price hike on weekends or monthends.

## Cancelled Vs Not Cancelled ADR

```python
cancelled_df_adr = df[df['is_canceled'] == 1].groupby('reservation_status_date')[['adr']].mean().reset_index()
not_cancelled_df_adr = df[df['is_canceled'] == 0].groupby('reservation_status_date')[['adr']].mean().reset_index()
```

```python
plt.figure(figsize=(20, 8))
plt.title('Average Daily Rate', size = 25)
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'Cancelled')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'Not Cancelled')
plt.legend(fontsize = 15)
plt.show()
```

The visualization looks messy as the data is widespread. We will filter the data to get a better visual.

```python
cancelled_df_adr = df[(df['is_canceled'] == 1) & (df['reservation_status_date'] > '2016')].groupby('reservation_status_date')[['adr']].mean().reset_index()
not_cancelled_df_adr = df[(df['is_canceled'] == 0) & (df['reservation_status_date'] > '2016')].groupby('reservation_status_date')[['adr']].mean().reset_index()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate', size = 25)
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'Cancelled')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'Cancelled')
plt.show()
```

### Conclusion
- Overall high cancellations when prices are high.
- The cancelled data is higher when compared to not cancelled data for the most part in terms of average daily rate.
- This proves that the average daily rate is heavily influencing the high cancellation rates.
- The spikes we observe are nothing but weekend or month end price hikes.

## Finding Cancellation Reason for Resorts
Is higher cencellations related to `lead_time` in resorts ?

```python
resort = df[df['hotel'] == 'Resort Hotel']
resort.head()
```

```python
resort_cancelled = resort[(resort['is_canceled'] == 1) & (resort['reservation_status_date'] > '2016')].groupby('reservation_status_date')[['lead_time']].mean().reset_index()
resort_not_cancelled = resort[(resort['is_canceled'] == 0) & (resort['reservation_status_date'] > '2016')].groupby('reservation_status_date')[['lead_time']].mean().reset_index()
```

```python
plt.figure(figsize=(20,5))
plt.title('Reservation Status W.R.T Lead Time', size = 25)
plt.plot(resort_cancelled['reservation_status_date'], resort_cancelled['lead_time'], label = 'Cancelled')
plt.plot(resort_not_cancelled['reservation_status_date'], resort_not_cancelled['lead_time'], label = 'Not Cancelled')
plt.ylabel('Lead Time', size=15)
plt.legend(fontsize = 10)
plt.show()
```

**Findings**

Longer lead time are leading to more cancellations in resort hotels

```python
cancelled_wl = resort[(resort['is_canceled'] == 1) & (resort['reservation_status_date'] > '2016')].groupby('reservation_status_date')[['days_in_waiting_list']].mean().reset_index()
not_cancelled_wl = resort[(resort['is_canceled'] == 0) & (resort['reservation_status_date'] > '2016')].groupby('reservation_status_date')[['days_in_waiting_list']].mean().reset_index()
```

```python
plt.figure(figsize=(20,6))
plt.title('Reservatioin Status W.R.T Waiting List', size = 25)
plt.plot(cancelled_wl['reservation_status_date'], cancelled_wl['days_in_waiting_list'], label = 'Cancelled')
plt.plot(not_cancelled_wl['reservation_status_date'], not_cancelled_wl['days_in_waiting_list'], label = 'Not Cancelled')
plt.legend(fontsize = 15)
plt.ylabel('Number of Days', size = 15)
plt.xlabel('Reservation Satus Date', size = 15)
plt.show()
```

### Findings
Longer waiting list has no result in higher cancellation rates

# üè®Summary and Conclusion

1. Strategic Pricing Adjustments: Cancellation rates tend to increase as room prices rise. To mitigate cancellations, hotels can optimize their pricing strategies by offering location-based discounts and special consumer incentives.

2. Weekend and Holiday Discounts: Resort hotels experience a higher cancellation-to-booking ratio compared to city hotels. To address this, hotels should consider offering competitive discounts on room rates during weekends and holidays.

3. January Revenue Boost Campaign: In the month of January, hotels can initiate revenue-boosting campaigns and marketing efforts at reasonable price points. This is especially valuable, given that January tends to have the highest cancellation rates.

4. Early Bird Benefits: Offer discounts or perks for guests who book in advance, encouraging early reservations.

5. Enhanced Quality and Service in Portugal: To reduce cancellation rates, hotels, particularly in Portugal, can focus on elevating the quality of their accommodations and services. This strategy aims to enhance guest satisfaction and loyalty,resulting in fewer cancellations.

6. Effective Cancellation Management: Recognizing the emerging trend of post-cancellation payments, viewed as an opportunity to turn challenges into advantages from a business standpoint, we can effectively leverage this situation. By employing astute strategies aimed at reducing cancellations, including transparent terms and conditions, considerate reminders, and potentially introducing incentives for guests committed to their reservations, we can create a win-win scenario. This approach not only enhances the guest experience but also guarantees a consistent revenue flow for the business.
